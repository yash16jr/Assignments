## Ans1

GET:
The GET method is used to retrieve data from a server. It is a request to the server to send data to the client. When a user submits a GET request, the browser sends a URL encoded query string to the server. This query string contains key-value pairs that represent the data being sent. GET requests are generally used to retrieve information that already exists on the server, such as a web page, image, or video. GET requests are also used to perform searches on a server.



POST:
The POST method is used to submit data to a server. Unlike GET, POST requests send data in the request body instead of the URL. This makes it possible to send much larger amounts of data. POST requests are generally used to create or update data on the server, such as adding a new user account or submitting a form.

## Ans2

Request is a built-in module in Flask that is used to handle HTTP requests sent by clients to the Flask application. The request module provides access to the data that is sent in the request, such as form data, query parameters, and headers.

## Ans3

The redirect() function in Flask is used to redirect the user to a different URL or endpoint within the same Flask application or to a different web application.


## Ans4

Templates in Flask are used to generate dynamic HTML pages that can be served to users in response to a request. Templates are written in HTML and include placeholders for dynamic content that is generated by the Flask application.

The render_template() function in Flask is used to render a template and generate an HTML page. This function takes a template filename as an argument and returns an HTML page generated from the template. The function looks for the template file in a folder called templates by default.

```python
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/hello', methods=['GET'])
def hello():
    return jsonify({'message': 'Hello, world!'})

if __name__ == '__main__':
    app.run(host='0.0.0.0')
```
