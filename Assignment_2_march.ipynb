## Ans1

Matplotlib is a Python library used for data visualization. It provides a wide variety of plotting tools for creating high-quality static, animated, and interactive visualizations in Python.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:-

Line plot - This is a plot where data is plotted as a series of points connected by lines.
Scatter plot - This is a plot where individual data points are plotted as points on a graph.
Bar plot - This is a plot where the data is presented as a series of bars.
Histogram - This is a plot that shows the distribution of data using bars of different heights.
Pie chart - This is a plot where the data is represented as a circular chart divided into slices.

## Ans2

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.title('Scatter plot of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Ans3

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Plotting the lines using subplot()
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

## Ans4

 A bar plot is a type of plot that represents categorical data with rectangular bars, where the height of each bar represents the magnitude of the variable being measured. Bar plots are used to compare the values of different categories or to show how a categorical variable changes over time.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company, profit)
plt.title('Bar plot of company and profit')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()
```

## Ans5

 A box plot, is a type of plot that displays the distribution of a dataset using quartiles. The box plot displays the range of the data, the median, and the interquartile range (IQR). It is used to visualize the spread and skewness of the data, and to identify outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
data = [box1, box2] # Data to be plotted
plt.boxplot(data) # Plotting the data
plt.title('Box Plot')
plt.xticks([1, 2], ['Box 1', 'Box 2']) 
plt.ylabel('Values')
plt.show()
```
